```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("Mall_Customers.csv")
print(dataset)
```

```python
X = dataset.iloc[:, 3].values
X = X.reshape(-1, 1)
print(X)
```

```python
y = dataset.iloc[:, 4].values
print(y)
```

```python
plt.scatter(X, y)
plt.xlabel("Salary")
plt.ylabel("Score")
plt.show()
```

```python


from sklearn.cluster import KMeans

wcss = []

X_original = dataset.iloc[:, [2,3, 4]].values

for i in range(1, 10):
    kmean = KMeans(n_clusters=i)
    kmean.fit(dataset.iloc[:, [3, 4]].values)
    wcss.append(kmean.inertia_)
    
plt.plot(range(1, 10), wcss)
plt.xlabel("Value of K")
plt.ylabel("Value of WCSS")
plt.show()
```

```python
print(X_original)
```

```python
kmeans = KMeans(n_clusters=5)
y_mean = kmeans.fit_predict(X_original)
print(y_mean)
```

```python
plt.scatter(X_original[y_mean == 0, 1], X_original[y_mean == 0, 0], s=5, c="red", label="One")
plt.scatter(X_original[y_mean == 1, 1], X_original[y_mean == 1, 0], s=5, c="green", label="Two")
plt.scatter(X_original[y_mean == 2, 1], X_original[y_mean == 2, 0], s=5, c="grey", label="Three")
plt.scatter(X_original[y_mean == 3, 1], X_original[y_mean == 3, 0], s=5, c="orange", label="Four")
plt.scatter(X_original[y_mean == 4, 1], X_original[y_mean == 4, 0], s=5, c="blue", label="Five")
plt.legend()
plt.show()
```
